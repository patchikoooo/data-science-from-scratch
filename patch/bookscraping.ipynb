```python
from bs4 import BeautifulSoup as bs4
import requests
import pandas as pd

url = "http://books.toscrape.com/catalogue/page-2.html"
#html = requests.get(url)


pages_to_scrape = 5
```

```python
pages = []
for i in range(1, pages_to_scrape + 1):
    url = ("http://books.toscrape.com/catalogue/page-{}.html").format(i)
    pages.append(url)
    #html = requests.get(url)
    print(url)
```

```python
bookshelf = []
imglinks = []
prices = []
stars = []
titles = []
urlss = []

for i in range(1, pages_to_scrape + 1):
    url = ("http://books.toscrape.com/catalogue/page-{}.html").format(i)
    html = requests.get(url).text
    soup = bs4(html, 'html.parser')
    bookshelf = soup.find('ol', {'class': 'row'})
    bookshelf = bookshelf('li', {'class': 'col-xs-6 col-sm-4 col-md-3 col-lg-3'})
    #print(shelf.prettify())
    for book in bookshelf:
        title = book.find('h3').getText()
        price = book.find('p', attrs={'class': 'price_color'}).getText()
        star = book.find('p').attrs.get('class')[1]
        link = 'http://books.toscrape.com/catalogue/' + book.find('a')['href']
        imglink = 'http://books.toscrape.com' + book.find('img')['src'][2::]
        
        titles.append(title)
        prices.append(price)
        stars.append(star)
        urlss.append(url)
        imglinks.append(imglink)
        #print(imglink)
        #price = price.encode('utf-8').decode('utf-8')
        #print(title, price, star)
        #print(link)
        
        #print("\n\n\n")
        #print(book.prettify())
```

```python
print(stars)
print(len(stars))
print(len(imglinks))
```

```python
data = {'Title': titles, 'Price': price, 'Rating': stars, 'Url': urlss, 'Imagelink': imglinks}
#data
```

```python
df = pd.DataFrame(data=data)
#df.reset_index(inplace=True)
df.index += 1   # increment the index
df
```
